```python

from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows
import pandas as pd

# Read data from report_A.xlsx using openpyxl
report_data = pd.read_excel(
    "C:\\Users\\XCG\\Desktop\\November\\Monthly_Reports\\Sapphire Bet (Hermes consulting company N.V.) October\\Sapphire_Bet_Hermes_Consulting_September_2023.xlsx",
    usecols=["TRANSACTION ID", "MERCHANT REFERENCE", "MERCHANT NAME", "APP NAME", "METHOD", "ORDER AMOUNT",
             "ORDER CURRENCY", "Processing Fee", "TAX", "CHARGEBACK FEE", "REFUND FEE", "AVS FEE", "STATUS",
             "ORDER TIME", "UPDATE TIME", "CONTENT", "SUBJECT"],dtype = {"TRANSACTION ID":object,"MERCHANT REFERENCE":object},
    parse_dates=["ORDER TIME", "UPDATE TIME"], sheet_name="Sheet-1"
)
## Filtering the specific currency
report_data = report_data[report_data["ORDER CURRENCY"]=="BRL"]
# Define the target file
output_file = "C:\\Users\\XCG\\Desktop\\November\\Monthly_Reports\\Sapphire Bet (Hermes consulting company N.V.) October\\Automated_report_final.xlsx"
sheet_name = "BRL Payin ORDERS"

# Load existing workbook or create a new one if not found
try:
    book = load_workbook(output_file)
except FileNotFoundError:
    book = openpyxl.Workbook()

# Remove the sheet if it already exists
if sheet_name in book.sheetnames:
    sheet = book[sheet_name]
    book.remove(sheet)

# Create a new sheet with the desired name
sheet = book.create_sheet(title=sheet_name)

# Write the new data to the sheet
for r_idx, row in enumerate(dataframe_to_rows(report_data, index=False, header=True), 1):
    for c_idx, value in enumerate(row, 1):
        sheet.cell(row=r_idx, column=c_idx, value=value)

# Save the changes
book.save(output_file)
```
