```python
## Sort homelessness by individuals
homelessness_sorted_ascending = homelessness.sort_values(["individuals"])
## Print the top few rows
homelessness_sorted_ascending.head(4)
```

```python
# Sort homelessness by descending family members
homelessness_sorted_descending = homelessness.sort_values(["individuals"], ascending=False)
homelessness_sorted_descending.head(4)
```

```python
# Sort homelessness by region, then descending family members
homelessness_by_region_fam_members = homelessness.sort_values(["region","family_members"], ascending = [True, False])
# or
homelessness_by_region_fam_members = homelessness.sort_values(["region","family_members"], ascending = [1, 0])
homelessness_by_region_fam_members.head(10)
```

```python
## # Select the individuals column
homelessness_indiv_ = homelessness["individuals"]
homelessness_indiv_.head()
```

```python
## Select the state and family_members columns
homelessness_state_fam_members = homelessness[["state","family_members"]]
homelessness_state_fam_members.head(4)
```

```python
# Select only the individuals and state columns, in that order
homelessness_individuals_state_ = homelessness[["individuals","state"]]
homelessness_individuals_state_.head()
```

```python
## Filter for rows where individuals is greater than 10000
homeless_greater_than_10k = homelessness[homelessness["individuals"] >= 10000]
homeless_greater_than_10k.head(5)
```
