```python
import pandas as pd
import numpy as np
```

```python
results_public = pd.read_csv("survey_results_public.csv")
results_public.head()
results_public.info()
```

```python
## How to drop Na from our Df?
results_public.dropna()
## As we can see, by default it deletes any row that has na value. We can conclude that all rows have at least one nan value
```

```python
# Let's say that I need to drop na from a specific column
results_public.dropna(axis="index", how="any", subset=["US_State","Sexuality"])
```

```python
# Imagine that some of our nan values are actually written as MIssing or NOOON and etc.
# We have to replace them and then drop them 
results_public["Gender"].unique()
```

```python
results_public.replace(['Non-binary, genderqueer, or gender non-conforming',
       'Man;Or, in your own words:',
       'Woman;Non-binary, genderqueer, or gender non-conforming',
       'Man;Non-binary, genderqueer, or gender non-conforming',
       'Or, in your own words:', 'Man;Woman',
       'Man;Woman;Non-binary, genderqueer, or gender non-conforming;Or, in your own words:',
       'Non-binary, genderqueer, or gender non-conforming;Or, in your own words:',
       'Woman;Or, in your own words:',
       'Man;Woman;Non-binary, genderqueer, or gender non-conforming',
       'Man;Non-binary, genderqueer, or gender non-conforming;Or, in your own words:',
       'Woman;Non-binary, genderqueer, or gender non-conforming;Or, in your own words:',
       'Man;Woman;Or, in your own words:'],np.nan,inplace=True)
```

```python
# I have converted all other genders and left only male and female. After converting them to nan, I have dropped them
results_public.dropna(axis="index", how="any", subset=["Gender"], inplace=True)
results_public
```

```python
## How to check whether a specific column has nan value
results_public["Gender"].isna() # As we can see, we have dropped all our nan values from this column
```
